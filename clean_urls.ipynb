```python
import pandas as pd
```

```python
top1m_df = pd.read_csv("data/urls/top-1m.csv")
```

```python
top1m_df.columns = ["ranked", "site"]
```

```python
top1m_df["lld"] = top1m_df.site.apply(lambda x: x.split(".")[0])
```

```python
top1m_df["tld"] = top1m_df.site.apply(lambda x: ".".join(x.split(".")[1:]))
```

```python
top1m_df.head(20)
```

```python
tld_order = top1m_df.tld.value_counts().reset_index()
```

```python
tld_order.columns = ['tld', 'count']
tld_order['rank'] = range(len(tld_order))
```

```python
tld_order.head()
```

```python
top10k = pd.DataFrame(columns=top1m_df.columns)
top10k

```

```python
for lld in top1m_df.lld.unique():
    if len(top10k) >= 10000:
        break
    top10k = top10k.append(top1m_df[top1m_df.lld == lld].iloc[0])
```

```python
top10k
```

```python
top10k = top10k.drop(["ranked", "lld", "tld"], axis=1)
top10k
```

```python
top10k.to_csv(path_or_buf="data/urls/top10k_cleaned.csv")
```
